```python
import pandas as pd
```

```python
perguntas = pd.read_csv('stackoverflow_perguntas.csv')
```

```python
perguntas.head()
```

```python
print(perguntas.shape)
```

```python
print(len(perguntas.Tags.unique()))
```

```python
print(perguntas.Tags.unique())
```

```python
label = list()
for tags in perguntas.Tags.unique():
  for tag in tags.split():
    if tag not in label:
      label.append(tag)

print(label)
```

```python
node_js = list()
for linha_tag in perguntas.Tags:
  if 'node.js' in linha_tag:
    node_js.append(1)
  else: node_js.append(0)

perguntas["node.js"] = node_js
```

```python
perguntas.head()
```

```python
def nova_coluna(lista_tags, dataframe, nome_tags):
  for tag in label:
    coluna = list()
    for linha_tag in dataframe[nome_tags]:
      if tag in linha_tag:
        coluna.append(1)
      else: coluna.append(0)
    dataframe[tag] = coluna
```

```python
nova_coluna(label,perguntas,"Tags")
```

```python
perguntas.head(10)
```

```python
lista_zip_tags = list(zip(perguntas[label[0]],
                     perguntas[label[1]],
                     perguntas[label[2]],
                     perguntas[label[3]]))

perguntas["todas_tags"] = lista_zip_tags
perguntas.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
perguntas_treino, perguntas_teste, tags_treino, tags_teste = train_test_split(
    perguntas.Perguntas,
    perguntas.todas_tags,
    test_size=0.2,
    random_state=123
)
```

```python
perguntas_treino
```

```python
perguntas_teste
```

```python
tags_treino
```

```python
tags_teste
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vetorizar = TfidfVectorizer(max_features=6000, max_df=0.85)
```

```python
vetorizar.fit(perguntas.Perguntas)
perguntas_treino_tfidf = vetorizar.transform(perguntas_treino)
perguntas_teste_tfidf = vetorizar.transform(perguntas_teste)
```

```python
import numpy as np
```

```python
tags_treino_asarray = np.asarray(list(tags_treino))
tags_teste_asarray = np.asarray(list(tags_teste))
```

#Relevância binária

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
regressao_logistica = LogisticRegression()
classificadoronevsrest = OneVsRestClassifier(regressao_logistica)
```

```python
classificadoronevsrest.fit(perguntas_treino_tfidf, tags_treino_asarray)
```

```python
resultado_onevsrest = classificadoronevsrest.score(perguntas_teste_tfidf, tags_teste_asarray)
print("Resultado {0: .2f}%".format(resultado_onevsrest*100))
```

```python
from sklearn.metrics import hamming_loss
```

```python
previsao_onevsrest = classificadoronevsrest.predict(perguntas_teste_tfidf)
hamming_loss_onevsrest = hamming_loss(tags_teste_asarray, previsao_onevsrest)
print("Hamming Loss {0: .2f}".format(hamming_loss_onevsrest))
```

#Classificação em cadeia

```python
from skmultilearn.problem_transform import ClassifierChain
```

```python
classificador_cadeia = ClassifierChain(regressao_logistica)
classificador_cadeia.fit(perguntas_treino_tfidf, tags_treino_asarray)
resultado_cadeia = classificador_cadeia.score(perguntas_teste_tfidf, tags_teste_asarray)
print("Resultado {0: .2f}%".format(resultado_cadeia*100))
previsao_cadeia = classificador_cadeia.predict(perguntas_teste_tfidf)
hamming_loss_cadeia = hamming_loss(tags_teste_asarray, previsao_cadeia)
print("Hamming Loss {0: .2f}".format(hamming_loss_cadeia))
```
